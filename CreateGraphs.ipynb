## Import libraries and define results data

Imports matplotlib and numpy, and defines dictionaries containing evaluation metrics for three models (Decision Tree, Random Forest, AdaBoost) on both global and language-specific tasks.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for ALL languages results
all_languages_data = {
    'Decision Tree': {'True F1': 0.24, 'False F1': 0.92, 'Macro Avg F1': 0.58, 'Accuracy': 0.86},
    'Random Forest': {'True F1': 0.26, 'False F1': 0.93, 'Macro Avg F1': 0.60, 'Accuracy': 0.88},
    'AdaBoost': {'True F1': 0.24, 'False F1': 0.92, 'Macro Avg F1': 0.58, 'Accuracy': 0.86}
}

# Data for Classification Report results
classification_report_data = {
    'Decision Tree': {'True F1': 0.21, 'False F1': 0.92, 'Macro Avg F1': 0.57, 'Accuracy': 0.86},
    'Random Forest': {'True F1': 0.24, 'False F1': 0.91, 'Macro Avg F1': 0.57, 'Accuracy': 0.84},
    'AdaBoost': {'True F1': 0.23, 'False F1': 0.90, 'Macro Avg F1': 0.57, 'Accuracy': 0.83}
}

def create_ml_graphs():
    # Create figure with subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Models for x-axis
    models = list(all_languages_data.keys())
    x_pos = np.arange(len(models))
    
    # Colors for different metrics
    colors = {
        'True F1': '#e74c3c',
        'False F1': '#27ae60', 
        'Macro Avg F1': '#3498db',
        'Accuracy': '#f39c12'
    }
    
    # Plot 1: ALL Languages Results
    ax1.set_title('Global (Giant) Model Results', fontsize=14, fontweight='bold', pad=20)
    
    for metric in ['True F1', 'False F1', 'Macro Avg F1', 'Accuracy']:
        values = [all_languages_data[model][metric] for model in models]
        ax1.plot(x_pos, values, marker='o', linewidth=2.5, markersize=8, 
                color=colors[metric], label=metric)
    
    ax1.set_xlabel('Models', fontsize=12)
    ax1.set_ylabel('Score', fontsize=12)
    ax1.set_xticks(x_pos)
    ax1.set_xticklabels(models, rotation=45, ha='right')
    ax1.set_ylim(0, 1)
    ax1.grid(True, alpha=0.3)
    ax1.legend(loc='upper left')
    
    # Format y-axis as percentage
    ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    
    # Plot 2: Classification Report Results
    ax2.set_title('Language-Specific Models Results', fontsize=14, fontweight='bold', pad=20)
    
    for metric in ['True F1', 'False F1', 'Macro Avg F1', 'Accuracy']:
        values = [classification_report_data[model][metric] for model in models]
        ax2.plot(x_pos, values, marker='o', linewidth=2.5, markersize=8, 
                color=colors[metric], label=metric)
    
    ax2.set_xlabel('Models', fontsize=12)
    ax2.set_ylabel('Score', fontsize=12)
    ax2.set_xticks(x_pos)
    ax2.set_xticklabels(models, rotation=45, ha='right')
    ax2.set_ylim(0, 1)
    ax2.grid(True, alpha=0.3)
    ax2.legend(loc='upper left')
    
    # Format y-axis as percentage
    ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    
    # Adjust layout and show
    plt.tight_layout()
    plt.show()

# Alternative function to create separate graphs
def create_separate_graphs():
    models = list(all_languages_data.keys())
    x_pos = np.arange(len(models))
    
    colors = {
        'True F1': '#e74c3c',
        'False F1': '#27ae60', 
        'Macro Avg F1': '#3498db',
        'Accuracy': '#f39c12'
    }
    
    # Graph 1: ALL Languages Results
    plt.figure(figsize=(10, 6))
    
    for metric in ['True F1', 'False F1', 'Macro Avg F1', 'Accuracy']:
        values = [all_languages_data[model][metric] for model in models]
        plt.plot(x_pos, values, marker='o', linewidth=2.5, markersize=8, 
                color=colors[metric], label=metric)
    
    plt.title('Global (Giant) Model Results', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Models', fontsize=12)
    plt.ylabel('Score', fontsize=12)
    plt.xticks(x_pos, models, rotation=45, ha='right')
    plt.ylim(0, 1)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    plt.tight_layout()
    plt.show()
    
    # Graph 2: Classification Report Results
    plt.figure(figsize=(10, 6))
    
    for metric in ['True F1', 'False F1', 'Macro Avg F1', 'Accuracy']:
        values = [classification_report_data[model][metric] for model in models]
        plt.plot(x_pos, values, marker='o', linewidth=2.5, markersize=8, 
                color=colors[metric], label=metric)
    
    plt.title('Language-Specific Models Results', fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Models', fontsize=12)
    plt.ylabel('Score', fontsize=12)
    plt.xticks(x_pos, models, rotation=45, ha='right')
    plt.ylim(0, 1)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))
    plt.tight_layout()
    plt.show()

# Run the functions
if __name__ == "__main__":
    # Create both graphs in one figure
    create_ml_graphs()
    
    # Or create separate graphs
    # create_separate_graphs()
```
